## Obtaining the Data

All data used here was obtained from the [Inside Airbnb](http://insideairbnb.com/get-the-data.html) website.

For this initial exploratory analysis, only the following file will be downloaded:

* `listings.csv` - *Summary information and metrics for listings in Barcelona (good for views).*

```
# importing needed packages
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```
# importing the listings.csv file for a DataFrame
df = pd.read_csv("http://data.insideairbnb.com/spain/catalonia/barcelona/2020-12-16/visualisations/listings.csv")
```

## Data analysis

This step aims to create an initial situational awareness and allow an understanding of how the data is structured.

**Dictionary of variables**

* `id` - id number generated to identify the property
* `name` - name of the advertised property
* `host_id` - id number of the owner (host) of the property
* `host_name` - Host name
* `neighborhood_group` - this column does not contain any valid values
* `neighborhood` - name of the neighborhood
* `latitude` - latitude coordinate of the property
* `longitude` - property longitude coordinate
* `room_type` - informs the type of room that is offered
* `price` - price to rent the property
* `minimum_nights` - minimum number of nights to book
* `number_of_reviews` - ​​number of reviews the property has
* `last_review` - date of the last review
* `reviews_per_month` - number of reviews per month
* `calculated_host_listings_count` - number of properties from the same host
* `availability_365` - number of availability days within 365 days

Before starting any analysis, let's check the face of our  *dataset*, analyzing the first 5 entries.

### **Q2. What is the percentage of missing values ​​in the *dataset*?**

### **Q3. What is the type of distribution of the variables?**

To identify the distribution of variables, I will plot the histogram.

```
# plot the histogram of numeric variables
df.hist(bins=15, figsize=(15,10));
```

### **Q4. Are there *outliers* present?**



By distributing the histogram, it is possible to verify the presence of *outliers*. Take for example the variables `price`,` minimum_nights` and `calculated_host_listings_count`.

Values ​​do not follow a distribution, and distort the entire graphic representation. To confirm, there are two quick ways to help detect *outliers*. Are they:

* Statistical summary using the `describe ()` method
* Plot `boxplots` for the variable.



```
# see the statistical summary of numerical variables
df[['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month',
    'calculated_host_listings_count', 'availability_365']].describe()
```

Looking at the statistical summary above, we can confirm some hypotheses such as:

* The `price` variable has 75% of the value below 100, but its maximum value is 11500.
* The minimum number of nights (`minimum_nights`) is over 365 days a year.

#### Boxplot for minimum_nights



```
# minimum_nights
df.minimum_nights.plot (kind = 'box', vert = False, figsize = (15, 3))
plt.show ()

# see amount of amounts over 30 days for minimum_nights
print("minimum_nights: valores acima de 30:")
print("{} entradas".format(len(df[df.minimum_nights > 30])))
print("{:.4f}%".format((len(df[df.minimum_nights > 30]) / df.shape[0])*100))
```

#### Boxplot for price

```
# price
df.price.plot(kind='box', vert=False, figsize=(15, 3),)
plt.show()

# see quantity of values ​​above 1500 for price
print("\nprice: valores acima de 1500")
print("{} entradas".format(len(df[df.price > 1500])))
print("{:.4f}%".format((len(df[df.price > 1500]) / df.shape[0])*100))

# df.price.plot(kind='box', vert=False, xlim=(0,1300), figsize=(15,3));
```

#### Histograms without *outliers*

Since we have identified *outliers* in the `price` and` minimum_nights` variables, we will now clear their *DataFrame* and plot the histogram again.

```
# remove * outliers * in a new DataFrame
df_clean = df.copy()
df_clean.drop(df_clean[df_clean.price > 1500].index, axis=0, inplace=True)
df_clean.drop(df_clean[df_clean.minimum_nights > 30].index, axis=0, inplace=True)

# remove `neighborhood_group` as it is empty
df_clean.drop('neighbourhood_group', axis=1, inplace=True)

# plot the histogram for numeric variables
df_clean.hist(bins=15, figsize=(15,10));
```

### **Q4. What is the correlation between the variables**

Correlation means that there is a relationship between two things. In our context, we are looking for a relationship or similarity between two variables.

This relationship can be measured, and it is a function of the correlation coefficient to establish its intensity. To identify the correlations between the variables of interest, I will:

* Create a correlation matrix
* Generate a *heatmap* from this matrix, using the `seaborn` library



```
# create a correlation matrix
corr = df_clean[['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month',
    'calculated_host_listings_count', 'availability_365']].corr()

display(corr)
```

```
sns.heatmap(corr, cmap='RdBu', fmt='.2f', square=True, linecolor='white', annot=True);
```

### **Q7. Word Cloud**

Here we can see the tag cloud for the most relevant words in Barcelona rental database, in other words it shows the most searched words in the Airbnb database acessed by the users and clients.

```
#install word
!pip install wordcloud -q
```

```
#import neeed packadge
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

#import the csv file for a dfv
df = pd.read_csv("/content/drive/MyDrive/listings.csv")

#eliminate columns with null values
description = df.dropna(subset=['description'], axis=0) ['description']
```

```
#example of proprerties description over airbnb
display(description.iloc[95])
display(description.iloc[111])
display(description.iloc[150])
```

```
#word concatenate
all_description = " ".join  (s for s in description)

#see quantity of words 
print("Quantity of words: {}".format(len(all_description)))
```

```
#stopwords
stopwords = set(STOPWORDS)
stopwords.update(["of", "the", "I", "you", "that", "this", "as", "these", "br", "en", "la", "con", "el", "number", "Exempt", "License", "number"])

#generate a wordcloud 
wordcloud = WordCloud(stopwords=stopwords,
                      background_color="black",
                      width=1600, height=800).generate(all_description)
                      
#Show the final image
fig, ax = plt.subplots(figsize=(10,6))
ax.imshow(wordcloud, interpolation='bilinear')
ax.set_axis_off()

plt.imshow(wordcloud);
wordcloud.to_file("airbnb_description_wordcloud.png")
```

```
# local address for your image
bcn_mask = np.array(Image.open("/content/drive/MyDrive/Barcelona_Colom2.jpg"))
   
# generate a wordcloud
wordcloud = WordCloud(stopwords=stopwords,
                      background_color="black",
                      width=2500, height=2500, max_words=2000,
                      mask=bcn_mask, max_font_size=200,
                      min_font_size=1).generate(all_description)

# show the final image
fig, ax = plt.subplots(figsize=(20,15))
ax.imshow(wordcloud, interpolation='bilinear')
ax.set_axis_off()

plt.imshow(wordcloud)
wordcloud.to_file("airbnb_description_wordcloud.png")
```
